**stop word **


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added to specifically download 'punkt_tab'

text = "I love machine learning and data science.it is so intresting subject for me."
tokens = word_tokenize(text.lower()) # Tokenize the text and convert to lowercase
remove_stop_word = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word not in remove_stop_word]
print(filtered_words)
```
