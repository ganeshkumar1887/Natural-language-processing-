Sentence tokenization


```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
text = "hello , I am ganesh .i am a software engineer."
```

```python
import nltk
nltk.download('punkt_tab')
sent = sent_tokenize(text)
sent
```

Word Tokenization (NLTK)


```python
from nltk.tokenize import word_tokenize
word = word_tokenize(text)
word
```

Character Tokenization

```python
characters = list(text)
print("Character Tokenization:")
print(characters)
```

Whitespace Tokenization

```python
text = "hello , I am ganesh .i am a software engineer."
tokens = text.split()
print("Whitespace Tokenization:")
print(tokens)
```

Regex Tokenization

```python
import re

text = "Email me at test123@gmail.com!"

tokens = re.findall(r'\b\w+\b', text)
print("Regex Tokenization:")
print(tokens)
```

Subword Tokenization – BPE (using HuggingFace)

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Tokenization is important for NLP"

tokens = tokenizer.tokenize(text)
print("Subword Tokenization (BERT):")
print(tokens)
```
